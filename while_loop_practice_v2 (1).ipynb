Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
#write answer here
i = 1
while i<=20:
  print(i)
  i+=1
```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
n = int(input("enter a number"))
i = 1
total = 0
while i<=n:
  total+= i
  print(f"sum : {total}")
  i+=1
```

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
#write answer here
n= int(input("enter a number"))
rev = 0
while n>0:
  digit = n%10
  rev = rev *10 + digit
  n//= 10
  print(f"reverse : { rev }")
```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
#write answer here
n = int(input("enter a number"))
count=0
while n>0:
  n//=10
  count+=1
print(f"count: {count}")
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here
n = int(input("enter a number: "))
temp=n
rev=0
while n>0:
  digit = n%10
  rev = rev*10 + digit
  n//=10

if temp==rev:
  print("palindrome")
else:
  print("not palindrome")
```

Q8. Guess the number game.

Approach:
- Use while loop
- Continue until correct guess

```python
#write answer here
fixed_no=10
n = int(input("enter a number"))
while n!=fixed_no:
  print("wrong guess")
  n= int(input("enter a number"))
print("correct guess")
print("end of program")
```
